We now understand the intuition behind both simple linear regression and multiple linear regression.  Recall that a simple linear regression problem can be modeled with the equation:

$$
y = b_0 + b_1x_1
$$

and that a multiple linear regression problem can be modeled with the equation:

$$
y = b_0 + b_1x_1 + b_2x_2 + ... b_nx_n
$$

These both model linear problems.  However, often a dataset may not be linear, and a straight line will not fit the data very well.  Observe the following figures:

```python
# import dependencies
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# define feature/label arrays
x = np.arange(1, 11)
y_simple = x*2
y_polynomial = np.array([1, 2, 4, 8, 16, 32, 64, 128, 256, 512])
```

```python
# Create figure
plt.figure(figsize=(15,5))

# Plot simple data
plt.subplot(1,2,1)
sns.regplot(x = x, y = y_simple, y_jitter=1)
plt.title('Figure 1. Simple Data Fitted With Simple Linear Regression')

# Plot polynomial data
plt.subplot(1,2,2)
sns.regplot(x = x, y = y_polynomial, y_jitter=1)
plt.title('Figure 2. Polynomial Data Fitted With Simple Linear Regression')
```

As seen in Figure 1, a straight line models the data well when we have simple linear data.  Figure 2 on the other hand demonstrates that a straight line is a poor fit to polynomial data.  When faced with polynomial data then, we may use polynomial regression.  A polynomial regression problem can be modeled with the following equation:


$$
y = b_0 + b_1x + b_2x^2 + ... b_nx^n
$$

Unlike with multiple linear regression, where we use different variables: $x_1, x_2, ... x_n$, we use the same variable raised to different powers: $x, x^2, x^3, ..., x^n$.  The following figure demonstrates the result of using polynomial regression.

```python
# Plot polynomial data
sns.regplot(x = x, y = y_polynomial, order=3, y_jitter = 50)
plt.title('Figure 3. Polynomial Data Fitted With Polynomial Regression')
```
