# Simple Linear Regression: Intuition

In this notebook, we discuss the intuition behind Simple Linear Regression, before diving into application.

Simple Linear Regression may be modeled with the formula

$$
y = b_0 + b_1 x_1
$$

Where $y$ is the dependent variable—or in machine learning, the label—$b_0$ is the constant, $b_1$ is the coefficient for $x_1$, and $x_1$, is the independent variable. Note that in simple linear regression, we have only one coefficient/constant pair; therefore it is not necessary to specifically label these with the numbers 0/1.  You may also be familiar with the formula:

$$
y = mx + b
$$

As we move into other types of linear regression, such as multiple linear regression, we will have to specifically label these with numbers; consider the multiple linear regression formula, as follows:

$$
y = b_0 + b_1 x_1 + b_2 x_2 + ... b_n x_n
$$

Turning our attention back to simple linear regression, let us now consider an example.  Suppose we want to predict a person's salary, based on their years of experience.  In this example, an individual with no experience starts at at least $50,000 per year.

```python
# Import dependencies
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

# Set matplotlib style
style.use('ggplot')
```

```python
# Define array with years of experience
years_of_experience = np.arange(start=0, stop=32, step=2)

# Define array with salaries
salaries = np.full(16, 50000) + years_of_experience*3000 + np.random.standard_normal(16) * 5000

# Define best-fit line
m,b = np.polyfit(years_of_experience, salaries, 1)
```

```python
# Create figure
plt.figure(figsize=(15,9))

# Plot actual datapoints
plt.scatter(x=years_of_experience
           ,y=salaries
           ,color='blue'
           ,label='Actual Salary')

# Plot best-fit line
plt.plot(years_of_experience,
         m*years_of_experience+b
        ,color='orange'
        ,label='Predicted Salary')

# Define title
plt.title('Salary Based on Years of Experience')

# Define labels
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Show legend
plt.legend()

# Show plot
plt.show()
print('Figure 1.')
```

The goal of simple linear regression is to determine the best-fit line, as illustrated in Figure 1.  Using this, we can take any input value (years of experience), and determine the output value (salary).  In Figure 1, the blue points illustrate a person's actual salary, while the orange line models what we would reasonably predict given their years of experience.
